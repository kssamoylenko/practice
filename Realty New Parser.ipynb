Первый вид списков для парсера — варианты формулировок для количества комнат в квартире. 
Остается проблема, из-за которой пасрер может ловить не все: в функции каждый запрос разбивается по пробелу и сравнивается со словарем, соответственно, объекты типа "1 ком" не ловятся, так как с таким образцом сравниваются отдельно "1" и отдельно "ком". Возможное решение — сравнивать весь запрос целиком, в таком случае будут возникать ошибки, если запрос выглядит как "сдать однокомнатную квартиру".

```python
def very_good_idea(request, patterns):
    rooms = []
    matchs = []
    for i in patterns:
        result = re.findall(i, request)
        if len(result) != 0:
            matchs.append('one_room')
    if len(matchs)!=0 :
            rooms.append('one_room') 
    return request, rooms
```

```python
def very_very_good(requests, patterns):
    return [very_good_idea(request, patterns) for request in requests]
```

```python
def very_complicated_idea(request):
    n_property = []
    rooms = []
    location = []
    center_location = []
    design = []
    warm = []
    
    matchs_ow = []
    matchs_re = []
    matchs_ro = []
    matchs_4ro = []
    matchs_loc = []
    matchs_des = []
    matchs_war = []
    
    for i in one_room:
        result = re.findall(i, request)
        if len(result) != 0:
            matchs_ro.append('one_room')
    if len(matchs_ro)!=0 :
            rooms.append('one_room') 
            
    for i in two_rooms:
        result = re.findall(i, request)
        if len(result) != 0:
            matchs_ro.append('two_room')
    if len(matchs_ro)!=0 :
            rooms.append('two_room')  
            
    for i in three_rooms:
        result = re.findall(i, request)
        if len(result) != 0:
            matchs_ro.append('three_rooms')
    if len(matchs_ro)!=0 :
            rooms.append('three_rooms')         
        
    for i in four_rooms:
        result = re.findall(i, request)
        if len(result) != 0:
            matchs_4ro.append('four_rooms')
    if len(matchs_4ro)!=0 :
            rooms.append('four_rooms')     
        
    for i in studio:
        result = re.findall(i, request)
        if len(result) != 0:
            matchs_ro.append('studio')
    if len(matchs_ro)!=0 :
            rooms.append('studio') 
    
    for i in own_property:
        result = re.findall(i, request)
        if len(result) != 0:
            matchs_ow.append('own_property')
    if len(matchs_ow)!=0 :
            n_property.append('own_property') 
    
    for i in rent:
        result = re.findall(i, request)
        if len(result) != 0:
            matchs_re.append('rent')
    if len(matchs_re)!=0 :
            n_property.append('rent') 
    
    for i in center_location:
        result = re.findall(i, request)
        if len(result) != 0:
            matchs_loc.append('center_location')
    if len(matchs_loc)!=0 :
            location.append('center_location') 
            
            
    for i in sea_location:
        result = re.findall(i, request)
        if len(result) != 0:
            matchs_loc.append('sea_location')
    if len(matchs_loc)!=0 :
            location.append('sea_location')   
            
    for i in warm:
        result = re.findall(i, request)
        if len(result) != 0:
            matchs_war.append('warm')
    if len(matchs_war)!=0 :
            warm.append('warm') 
            
            
    for i in design:
        result = re.findall(i, request)
        if len(result) != 0:
            matchs_des.append('design')
    if len(matchs_des)!=0 :
            design.append('design')                 
    
    
    return request, n_property, rooms, location, warm, design 
```

```python
def very_very_complicated(requests):
    return [very_complicated_idea(request) for request in requests]
```

```python
super_complicated = very_very_complicated(realty['request'])
```
